```python
 #questao 1
 #define a função com os parâmetros indicados nas próximas linhas de comando
 def transformar_numeros (numero): 
     #inclui mais dados da tabela indicada, pois não seria suficiente para compor todos os números de 1 a 3999
     #conforme indicação da tabela de conversão de números romanos, encontrada em:
     #https://fazaconta.com/calculadora-romanos.htm
     #uso da função split para string, conforme indicado material Aula 07
     #o split retorna uma lista das palavras na string
     numeros_arabicos = '1000, 900, 500, 400, 100, 90, 50, 40, 10, 9, 5, 4, 1'.split(', ') 
     numeros_romanos = 'M', 'CM', 'D', 'CD', 'C', 'XC', 'L', 'XL', 'X', 'IX', 'V', 'IV', 'I'.split(', ')
     #faz a leitura dos dados do split
     resultado = ''
     #execução em loop 
     for i in range(len(numeros_arabicos)):
       while numero >= int(numeros_arabicos[i]):
             numero -= int(numeros_arabicos[i])
             resultado += str(numeros_romanos[i]).replace('['', '').replace('']', '')
     #retorna o resultado da conversão
     return resultado

while True:
    try:
        #mensagem para deixar o usuário finalizar o programa
        print('Se quiser finalizar o programa digite 0 (zero).')
        #entrada de dados pelo usuário
        arabico_input = int(input('Digite o numero para conversão: '))
        # opção zero para sair do programa
        if arabico_input == 0:
            #mensagem para finalização do programa
            print('Finalizando o programa!')
            #função que encerra o loop
            break
        #permite input até o 3999
        elif arabico_input > 3999:
            #mensagem para caracter inválido
            print("Valor acima do permitido (1-3999). Tente novamente")
        else:
            #conversão dos números
            romano_output = transformar_numeros (arabico_input)
            #saída para o usuário do resultado da conversão
            print('Número arábico {0} corresponde ao número romano:{1}'.format(arabico_input, romano_output))
    #interrupção gerada pelo usuário
    except:
      #mensagem para caracter inválido
      print(f"Você digitou um caracter inválido. Tente novamente.")
```

```python
#questao 3
#uso def para usar função que retorna o digito verificador da conta
def digito_verificador(num):
    #tranforma o numero em string para inverter
    num_str = str(num)
    num_inv = num_str[::-1]
    #faz a soma do número e seu inverso
    soma = num + int(num_inv)
    #tranforma a soma em string para o fatiamento necessário
    soma_str = str(soma)
    #cria a variavel para realizar a soma das multiplicações do número
    soma2 = 0
    #uso for loop para fazer o fatiamento e a multiplicação
    for i in range(len(soma_str)):
        soma2 += int(soma_str[i])*(i+1)
    #transforma a soma2 em string para pegar o ultimo caracter
    soma2_str = str(soma2)
    #retorna o ultimo digito como valor numerico
    return int(soma2_str[-1])
#leitura das instruções enquanto a condição for verdadeira
while True:
    try:
        #entrada do texto a ser analisado, sem definir o tipo para usar também como string
        num_conta = input('Informe o número da conta corrente com 5 dígitos: ')
        #analisa a quantidade de caracteres do texto
        len_num_conta = len(num_conta)
        #tranforma em integer 
        conta = int(num_conta)
        #se o texto for maior que 5 caracteres, retorna advertência ao usuário
        if len_num_conta < 5:
            print(f'Vocẽ digitou um número com a quantidade de dígitos menor do que a necessária')
            print('Tente novamente.')
        #se o texto for menor que 5 caracteres, retorna advertência ao usuário
        elif len_num_conta > 5:
            print(f'Vocẽ digitou um número com a quantidade de dígitos maior do que a necessária')
            print('Tente novamente.')
        #se o texto estiver correto, imprime as informações necessárias
        elif len_num_conta == 5:
            print(f'O número da sua conta com digíto verificador é: {conta}-{digito_verificador(conta)}')
            print('Programa finalizado.')
            #parar o programa
            break
    #trata os erros digitados pelo usuário
    except:
        #se não digitado um caracter numerico (isnumeric), retorna advertência ao usuário
        if not num_conta.isnumeric():
            print(f'Voce digitou um caracter não permitido: {num_conta}')
            print('Recomece o programa novamente')   
            #break
```
